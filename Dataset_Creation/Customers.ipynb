```python
import pandas as pd
```

```python
dfl = pd.DataFrame()
lst = []
dfl = pd.read_csv('address.csv')
lst = dfl['1671 Lauren Knoll, Suite 945, 9012, Lake Audreyborough, Tasmania, Australia'].to_list()
lst.append('1671 Lauren Knoll, Suite 945, 9012, Lake Audreyborough, Tasmania, Australia')
len(lst)
```

```python
df = pd.DataFrame()
df['address'] = lst
df
```

```python
def split(t):
  add = t.split(',')
  home_address = add[0].strip() + add[1].strip()
  zip_code = add[2].strip()
  city = add[3].strip()
  state = add[4].strip()
  country = add[5].strip()
  return home_address, zip_code, city, state, country

home_address_list = []
zip_code_list = []
city_list = []
state_list = []
country_list = []

for idx, row in df.iterrows():
    home_address, zip_code, city, state, country = split(row.address)
    home_address_list.append(home_address)
    zip_code_list.append(zip_code)
    city_list.append(city)
    state_list.append(state)
    country_list.append(country)

df['home_address'] = home_address_list
df['zip_code'] = zip_code_list
df['city'] = city_list
df['state'] = state_list
df['country'] = country_list
```

```python
df_final = pd.merge(df_c, df, left_index=True, right_index=True)
df_final = df_final.drop(['address'], axis=1)
df_final
```
