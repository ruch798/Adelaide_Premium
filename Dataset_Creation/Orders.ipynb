```python
import pandas as pd
import random
from datetime import datetime
```

```python
df = pd.read_csv("MOCK_DATA.csv")
df
```

```python
def add_days(d):
  i = random.randrange(1,28)
  dd = ((pd.to_datetime(d) + pd.DateOffset(days=i)).to_pydatetime().date()).strftime('%m/%d/%Y')
  return dd
```

```python
delivery_date = []
for idx, row in df.iterrows():
  dd = add_days(row.order_date)
  delivery_date.append(dd)

df['delivery_date'] = delivery_date
```
