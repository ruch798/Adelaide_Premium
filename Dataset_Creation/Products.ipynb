```python
import pandas as pd
import random
```

```python
shirts = ["Oxford Cloth", "Dress", "Flannel", "Denim", "Linen", "Camp Collared","Chambray", "Polo", "Cuban Collar", "Mandarin Collar", "Henley", "Casual Slim Fit"]
jackets = ["Denim", "Puffer", "Windbreaker", "Leather", "Bomber", "Cardigan", "Coach", "Parka", "Trench Coat", "Peacoat", "Pullover", "Shearling"]
trousers = ["Chinos", "Cords", "Drawstring", "Slim-Fit", "Wool", "Relaxed Leg", "Joggers", "Cropped", "Cargo Pants", "High-Waisted", "Pleated", "Tracksuit Bottoms"]
```

```python
colours = ["red", "orange", "yellow", "green", "blue", "indigo", "violet"]
sizes = ["XS", "S", "M", "L", "XL"]
```

```python
def create_df(item_names):
  prices = random.sample(range(90, 120), 12)
  
  df = pd.DataFrame()

  df['product_name'] = item_names
  df['price'] = prices

  df['size'] = [sizes for i in df.index]
  df['colour'] = [colours for i in df.index]

  df = df.explode('colour')
  df = df.explode('size')

  return df
```

```python
df_shirts = create_df(shirts)
df_shirts['product_type'] = "Shirt"

df_trousers = create_df(trousers)
df_trousers['product_type'] = "Trousers"

df_jackets = create_df(jackets)
df_jackets['product_type'] = "Jacket"
```

```python
df = pd.concat([df_shirts,df_jackets,df_trousers])

df['quantity'] = [random.randint(40,80) for x in range(len(df))]

df.insert(0, 'product_ID', range(len(df)))
df.reset_index(inplace = True, drop = True)

col = df.pop("product_type")
df.insert(1, col.name, col)

col2 = df.pop("price")
df.insert(5, col2.name, col2)

df
```

```python
def desc(color, size, name, product_type):
  desc = "A " + color + " coloured, " + size + " sized, " + name + " " + product_type
  return desc

des = []
for idx, row in df.iterrows():
    d = desc(row['colour'], row['size'],row['product_name'],row['product_type'])
    des.append(d)
```

```python
df['description'] = des
df
```

```python
df.to_csv("products.csv")
```
