```python
sales = pd.DataFrame()
sales['order_id'] = [random.randint(1,1000) for x in range(5000)]
sales['product_id'] = [random.randint(1,1259) for x in range(5000)]

sales
```

```python
sales.order_id.value_counts()
```

```python
sales.product_id.value_counts()
```

```python
sales = sales.merge(products[['product_id', 'price']])
sales.rename(columns = {'price':'price_per_unit'}, inplace = True)
sales
```

```python
sales = sales.sort_values('order_id')
sales
```

```python
sales['quantity'] = [random.randint(1,3) for x in range(5000)]
sales['total_price'] = sales['quantity'] * sales['price_per_unit']
```

```python
sales.insert(0, 'sales_id', range(len(sales)))
sales.reset_index(inplace = True, drop = True)
sales
```

```python
sales.to_csv("sales.csv")
```
